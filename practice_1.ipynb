```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv('https://raw.githubusercontent.com/ine-rmotr-curriculum/FreeCodeCamp-Pandas-Real-Life-Example/master/data/sales_data.csv',
    parse_dates=['Date'])
```

```python
sales.head(2)
```

```python
sales.Customer_Age.mean().round(1)
```

```python
sns.set_style("darkgrid")
fig, axs = plt.subplots(ncols = 2, figsize = (15,5))
ax0 = sns.kdeplot(x = "Customer_Age", data = sales, ax = axs[0])
ax0.set_title("Customer age kde")
ax1 = sns.boxplot(x = "Customer_Age", data = sales, ax = axs[1])
ax1.set_title("Customer age box plot")
plt.show()
```

```python
sales.Order_Quantity.mean()
```

```python
fig, axs = plt.subplots(ncols = 2, figsize = (14, 5))
ax0 = sns.histplot(x = "Order_Quantity", data = sales, ax = axs[0], bins = 30,)
ax0.set_title("order quantity histogram")
ax1 = sns.boxplot(x = "Order_Quantity", data = sales, ax = axs[1])
ax1.set_title("Order_Quantity box plot")
plt.show()
```
